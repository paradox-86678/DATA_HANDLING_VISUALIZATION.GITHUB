```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
plt.pie(x)
plt.show
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
y=["cherries","Apple","Banana","Mango"]

plt.pie(x, labels= y)
plt.show
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
y=["cherries","Apple","Banana","Mango"]
z=["blue","pink","black","green"]
plt.pie(x, labels= y, colors=z)
plt.show
```

```python

import matplotlib.pyplot as plt
x=[20,30,40,50]
y=["cherries","Apple","Banana","Mango"]
z=[0.2,0,0,0]
plt.pie(x, labels= y ,explode=z)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
y=["cherries","Apple","Banana","Mango"]
z=[0.2,0,0,0]
plt.pie(x, labels= y ,explode=z,shadow="true")

plt.show()
```

```python
x=(["A","B","C","D"])
y=([2,5,1,10])
plt.scatter(x,y,color="r")
plt.show()
```

```python
x=([1,2,3,4])
y=([1,2,3,4])
z=(["pink","blue","black","red"])
sizes=[200,300,480,380]
plt.scatter(x,y,color=z, marker="*", edgecolor="green",s=sizes,alpha=1)
plt.title("Scatter graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
g=[1,2,3,4]
h=[4,5,6,7]

plt.subplot(2,2,1)

plt.plot(g,h)
plt.subplot(2,2,2)

a=[20,30,40,50]
b=["cherries","Apple","Banana","Mango"]
plt.pie(a, labels= b)
plt.subplot(2,2,3)

x=(["A","B","C","D"])
y=([2,5,1,10])
plt.bar(x,y)
plt.subplot(2,2,4)

d=([1,2,3,4])
e=([1,2,3,4])
z=(["pink","blue","black","red"])
sizes=[200,300,480,380]
plt.scatter(d,e,color=z, marker="*", edgecolor="green",s=sizes,alpha=1)
plt.title("Scatter graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
