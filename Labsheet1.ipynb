```python
#Pie Chart
x=[10,20,45,15,10]
plt.pie(x)
plt.title('Values')
plt.show()
```

```python
x=[100,200,450,150,100]
lb=['A','B','C','D','E']
plt.pie(x,labels=lb)
plt.title('Values')
plt.show()
```

```python
clrs=["blue","orange","lightgreen","#0FBBFF","#00FFFF"]
plt.pie(x,labels=lb,colors=clrs)
plt.show()
```

```python
plt.pie(x,labels=lb,colors=clrs,explode=[0.05,0,0.2,0,0])
```

```python
#Bar Graphs
#Vertical
plt.bar(lb,x,color=clrs)
plt.show()
```

```python
#Hoizontal
plt.barh(lb,x,color=clrs)
plt.show()
```

```python
a=[1,2,3]
b=[3,2,2]
plt.subplot(2,1,1)
plt.plot(a,b)
plt.title("1st Subplot")
plt.subplot(2,1,2)
plt.plot(a,b)
plt.title("2nd Subplot")
plt.show()
```

```python
plt.plot(lb,x,)
plt.show()
```

```python
#Scatter_Plot
plt.scatter(lb,x,c=clrs)
plt.show()
```

```python
plt.scatter(lb,x,c=clrs,marker='*',edgecolors='b',s=x)
plt.show()
```

```python
plt.scatter(lb,x,marker='*',edgecolors='green',s=x,alpha=0.7)
plt.show()
```
